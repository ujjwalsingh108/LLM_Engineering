## CHAT COMPLETIONS API MANUAL

```python
import os
from dotenv import load_dotenv
from IPython.display import Markdown, display
```

```python
# Load environment variables from .env file
load_dotenv(override=True)
api_key = os.getenv("OLLAMA_API_KEY")
api_model = os.getenv("OLLAMA_MODEL")
base_url = os.getenv("OLLAMA_BASE_URL")

# Check if the API key and model are set
if not api_key or not api_model or not base_url:
    raise ValueError("Please set the OLLAMA_API_KEY, OLLAMA_MODEL, and OLLAMA_BASE_URL environment variables in your .env file.")
elif not api_key.startswith("ollama") or not api_model.startswith("llama"):
    raise ValueError("The OLLAMA_API_KEY/OLLAMA_MODEL should start with 'ollama'/'llama'. Please check your .env file.")
else:
    print("Environment variables loaded successfully.")
```

```python
import requests

headers = {"Authorization": f"Bearer {api_key}", "Content-Type": "application/json"}

payload = {
    "model": api_model,
    "messages": [
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "What is the capital of France?"}
    ]
}

print("Sending request to Ollama API with payload:\n", payload)
```

```python
response = requests.post(f"{base_url}/chat/completions", headers=headers, json=payload)

print(response.json())
```

```python
print(response.json()["choices"][0]["message"]["content"])
```

## CHAT COMPLETIONS API USING OPENAI SDK AS A WRAPPER

```python
from openai import OpenAI
client = OpenAI(base_url=base_url, api_key=api_key)

print("Sending request to Ollama API using OpenAI client:", client)
```

```python
# Get a fun fact about India

res = client.chat.completions.create(
    model=api_model,
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "Give me a fun fact about India."}
    ]
)
print(res.choices[0].message.content)
```

```python
res1 = client.chat.completions.create(
    model=api_model,
    messages=[
        {"role": "system", "content":"You are an expert in Mathematics and also a college professor."},
        {"role":"user", "content":"Explain the concept of derivatives in calculus."}
    ]
)

print(res1.choices[0].message.content)
```

```python
res2 = client.chat.completions.create(
    model= api_model,
    messages=[
        {"role": "system", "content": "You are a snarky assistant."},
        {"role": "user", "content": f"How much is {2 + 8} ?"}
    ]
)

print(res2.choices[0].message.content)
```

```python
system_prompt = """
You're a helpful assistant that provides concise and accurate answers to user queries. 
Always respond in a friendly and professional manner, and ensure that your answers are easy to understand. 
If you don't know the answer to a question, it's okay to say that you don't know, but try to provide some guidance on where the user might find the information they're looking for.
"""
```

```python
user_prompt = "What is the tallest mountain in the world?"
```

```python
def create_messages_dictionary(param):
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": f"{user_prompt} {param}"}
    ]
    return messages
```

```python
print(create_messages_dictionary("Please provide the answer in a single sentence."))   
```

```python
res3 = client.chat.completions.create(
    model=api_model,
    messages=create_messages_dictionary("Please provide the answer in a single sentence.")
)

print(res3.choices[0].message.content.split(".")[0] + ".")
```

## FIRST PROJECT (SCRAPPING A WEBSITE)

```python
from scraper import fetch_page_content

webpage_content = fetch_page_content("https://cnn.com")
print(webpage_content)
```

```python
system_prompt = """
You are a helpful assistant that analyzes the contents of a webpage and provides a long, detailed summary of its content. 
Respond in markdown format, and use bullet points to organize the information.
"""
```

```python
user_prompt = """
Here are the contents of a webpage.
Provide a detailed summary of the content in markdown format, using bullet points to organize the information.
If the content is too long, summarize only the most important points.
"""
```

```python
def summarize():
    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt + "\n\n" + webpage_content }
    ]
    response = client.chat.completions.create(
        model=api_model,       
        messages=messages
    )
    return response.choices[0].message.content
```

```python
def display_nicely():
    summary = summarize()
    display(Markdown(summary))
```

```python
display_nicely()
```
